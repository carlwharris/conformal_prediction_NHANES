```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
n = 250
alpha = 0.2
l = np.floor((n+1)*alpha)
x = np.linspace(beta.ppf(0.01, n+1-l, l), beta.ppf(0.99, n+1-l, l), 100)
plt.plot(x, beta.pdf(x, n+1-l, l),'r-')
plt.show()
```

```python
epsilon = 0.05
beta.cdf((1 - alpha) + epsilon, n+1-l, l) - beta.cdf((1 - alpha) - epsilon, n+1-l, l)
```

```python
beta.cdf((1 - alpha) - epsilon, n+1-l, l)
```

```python
beta.cdf(0.96, n+1-l, l)
```
